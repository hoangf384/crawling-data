```python
def changing_id(df):
    if 'data_id' in df.columns:
        df = df.rename(columns={'data_id': 'pid'})
    if 'pid' in df.columns:
        df = df.set_index('pid')
    return df

def return_list(x):
    if isinstance(x, str) and x.startswith('[') and x.endswith(']'):
        try:
            x = ast.literal_eval(x)
        except:
            pass
    return x
```

```python
def highspeed_ports(x):
    """
    Nhận input là string, list hoặc pandas Series.
    Trả về True nếu phát hiện Thunderbolt 3-5 hoặc USB 4.
    """
    pattern = re.compile(r"(thunderbolt\s*[3-5]|usb\s*4)", flags=re.IGNORECASE)

    def check_item(item):
        # nếu là list thì check từng phần
        if isinstance(item, list):
            for sub in item:
                parts = re.split(r"[\/,]", str(sub))
                for part in parts:
                    if pattern.search(part):
                        return True
        else:  # string hoặc các kiểu khác
            parts = re.split(r"[\/,]", str(item))
            for part in parts:
                if pattern.search(part):
                    return True
        return False

    # nếu là Series thì apply
    if hasattr(x, "apply"):
        return x.apply(check_item)
    else:
        return check_item(x)
```

```python
gate["Công nghệ Thunderbolt"] = gate["Cổng giao tiếp"].apply(highspeed_ports)
```
