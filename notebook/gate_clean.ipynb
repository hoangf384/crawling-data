```python
def highspeed_ports(x):
    """
    Kiểm tra sự xuất hiện của cổng tốc độ cao (Thunderbolt 3-5 hoặc USB 4).

    Hàm này nhận vào một chuỗi hoặc danh sách, sau đó tìm kiếm
    các cổng kết nối tốc độ cao (Thunderbolt từ 3 đến 5, hoặc USB 4). 
    Hỗ trợ nhiều định dạng như: phân tách bằng dấu "/" hoặc ",".

    Tham số
    ----------
    x : str hoặc list
        - Nếu là `str`: kiểm tra trực tiếp chuỗi.
        - Nếu là `list`: kiểm tra từng phần tử trong danh sách.

    Kết quả trả về
    ----------
    bool
        - Trả về `True` nếu phát hiện "Thunderbolt 3", "Thunderbolt 4", 
          "Thunderbolt 5" hoặc "USB 4".
        - Trả về `False` nếu không phát hiện.
    """
    pattern = re.compile(r"(thunderbolt\s*[3-5]|usb\s*4)", flags=re.IGNORECASE)

    if isinstance(x, list):
        for sub in x:
            parts = re.split(r"[\/,]", str(sub))
            for part in parts:
                if pattern.search(part):
                    return True
    else:  # string hoặc kiểu khác
        parts = re.split(r"[\/,]", str(x))
        for part in parts:
            if pattern.search(part):
                return True
    return False
```
