```python
# rule
## Công nghệ CPU
## tách ra tên hãng: AMD, Intel, Apple
## model: a,b = split(-); model = b
```

```python
def extract_cpu_str(x):
    if not isinstance(x, str):
        return {"Brand": None, "Model": None}
    
    # --- Hãng ---
    if any(k in x.lower() for k in ["intel", "celeron", "pentium"]):
        brand = "Intel"
    elif any(k in x.lower() for k in ["amd", "athlon", "ryzen"]):
        brand = "AMD"
    elif "apple" in x.lower() or "m1" in x.lower() or "m2" in x.lower() or "m3" in x.lower():
        brand = "Apple"
    elif "snapdragon" in x.lower():
        brand = "Snapdragon"
    else:
        brand = None
    
    # --- Model ---
    model = None
    
    if brand == "Apple":
        if "-" in x:
            model = x.split("-")[0].strip()
        else:
            model = x.strip()
        model = model.replace("Apple", "").strip().capitalize()
    elif "-" in x:
        if brand == "Snapdragon":
            model = x.split("-", 1)[1].strip()
            model = model.replace(" ","-")
        else:
            # Intel / AMD => lấy phần sau dấu "-"
           model = x.split("-")[1].strip()
    # else:
    #     # fallback regex nếu không có dấu "-"
    #     match = re.search(r"\b([A-Z]?\d{3,5}[A-Z]{0,3}|M\d(\s+\w+)*)$", x.strip(), flags=re.I)
    #     if match:
    #         model = match.group(1).strip()
    
    return {"Brand": brand, "Model": model}
```

```python
chip[["Brand", "Model"]] = chip["Công nghệ CPU"].apply(extract_cpu_str).apply(pd.Series)
```
