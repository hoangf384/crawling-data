```python
def split_storage(x):
    """
    Tách và chuẩn hóa dữ liệu về dung lượng lưu trữ.

    Parameters
    ----------
    x : list, str, or None
        Thông tin lưu trữ đầu vào. Thường là list (dung lượng + hỗ trợ nâng cấp).

    Returns
    -------
    pandas.Series
        Series với 3 cột: ["Dung lượng", "tmp1", "tmp2"].
    """
    
    if isinstance(x, list):
        # đảm bảo độ dài 3, thiếu điền None
        return pd.Series([x[i] if i < len(x) else None for i in range(3)],
                         index=["Dung lượng", "tmp1", "tmp2"])
    else:
        return pd.Series([x, None, None], index=["Dung lượng", "tmp1", "tmp2"])
```

```python
def extract_n_swap(row, col="Dung lượng"):
    """
    Chuẩn hóa dữ liệu lưu trữ bằng cách hoán đổi và trích xuất thông tin.

    Parameters
    ----------
    row : pandas.Series
        Một dòng dữ liệu từ DataFrame.
    col : str, default "Dung lượng"
        Tên cột chính để xử lý.

    Returns
    -------
    pandas.Series
        Dòng dữ liệu đã được cập nhật:
        - col : str
            Giá trị dung lượng đã làm sạch.
        - tmp1 : str
            Giá trị "hỗ trợ" (nếu có).
        - tmp3 : str
            Thông tin "(Có thể tháo...)" (nếu có).
    """
    x = row[col]

    # --- 1. swap
    if isinstance(x, str) and "hỗ trợ" in x.lower():
        row[col], row["tmp1"] = row["tmp1"], x
        x = row[col]  

    # --- 2. extract
    if isinstance(x, str):
        match = re.search(r"\(có thể tháo.*$", x, flags=re.I)
        value = match.group(0) if match else None
        clean = re.sub(r"\(có thể tháo.*$", "", x, flags=re.I).strip()

    # -- 3. update
        row[col] = clean
        row["tmp3"] = value

    return row
```

```python
def clean_tmp3(x):
    """
    Trích xuất thông tin từ chuỗi trong cột tmp3.

    Parameters
    ----------
    x : str or any
        Giá trị đầu vào, thường là nội dung từ cột tmp3.

    Returns
    -------
    str or None
        Chuỗi trích xuất được bắt đầu từ ký tự số đầu tiên
        cho đến trước ký tự cuối cùng. Nếu không tìm thấy
        hoặc đầu vào không phải string thì trả về None.
    """
    if not isinstance(x, str):
        return None
    match = re.search(r"\d.*(?=.$)", x, flags=re.IGNORECASE)
    if match:
        return match.group(0)
    return None
```

```python
def swap(row):
    """
    Chuẩn hóa vị trí SSD và HDD từ các cột tmp1 và tmp2.

    Parameters
    ----------
    row : pandas.Series
        Một dòng dữ liệu có các cột:
        - tmp1 : str
            Ổ cứng chính.
        - tmp2 : str
            Ổ cứng phụ.

    Returns
    -------
    tuple of (str or None, str or None)
        - Phần tử 1: SSD (nếu có).
        - Phần tử 2: HDD (nếu có).
        Trả về (None, None) nếu không xác định được.
    """
    x, y = row["tmp1"], row["tmp2"]    
    if isinstance(x, str) and "ssd" in x.lower():
        return x, y
    elif isinstance(x, str) and "hdd" in x.lower():
        return y, x
    return None, None
```

```python
storage[["tmp1", "tmp2"]] = storage[["tmp1", "tmp2"]].apply(swap, axis=1).apply(pd.Series)
storage.tmp1.value_counts()
```

```python
def clean_tmp1(x):
    """
    Chuẩn hóa thông tin ổ cứng từ tmp1.

    Parameters
    ----------
    x : str or any
        Giá trị từ cột tmp1. Có thể là chuỗi mô tả ổ cứng hoặc kiểu dữ liệu khác.

    Returns
    -------
    tuple
        (value, note)
        - value : str
            Dung lượng ổ cứng đã chuẩn hóa (ví dụ "1 TB", "2 TB").
            Nếu không tìm thấy thì trả về "không công bố".
            Nếu x không phải string thì trả về chính x.
        - note : str or None
            Ghi chú bổ sung, ví dụ `"gen 5 supported"` nếu có.
    """
    if not isinstance(x, str):
        return x, None  # (value, note)
    note = None
    if "gen 5" in x.lower():
        note = "gen 5 supported"
    
    m = re.search(r"(\d+\s*TB)", x, flags=re.I)
    if m:
        start = m.start()

        if x.endswith(")"):
            val = x[start:-1]
        else:
            val = x[start:]
        return val.strip(), note
    else:
        return "không công bố", note
```

```python
storage[["tmp1", "Note"]] = storage["tmp1"].apply(
    lambda x: pd.Series(clean_tmp1(x), index=["capacity", "Note"])
)
```

```python
def clean_tmp2(val, idx=None):
    """
    Chuẩn hóa thông tin ổ cứng từ cột tmp2.

    Parameters
    ----------
    val : str, float, or None
        Giá trị từ cột tmp2. Có thể là chuỗi mô tả, NaN hoặc None.
    idx : int, optional
        Chỉ số dòng. 

    Returns
    -------
    str or None
        - Nếu tìm thấy dung lượng dạng "X TB" trong chuỗi, trả về giá trị đó (ví dụ "2 TB").
        - Nếu là chuỗi nhưng không có thông tin dung lượng, trả về "không công bố".
        - Nếu là None, NaN hoặc kiểu dữ liệu khác, trả về None.
    """
    if pd.isna(val):
        return None
    
    if isinstance(val, str):
        match = re.search(r"(\d+\s*TB)", val, flags=re.I)
        if match:
            return match.group(0)  
        else:
            return "không công bố"

    return None
```

```python
storage["tmp2"] = storage.apply(
    lambda row: clean_tmp2(row["tmp2"], idx=row.name), axis=1
)
```
