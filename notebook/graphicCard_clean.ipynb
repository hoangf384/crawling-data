````python
def classify(x):
    """
    Phân tách chuỗi thành hai phần dựa trên dấu " - " và loại bỏ phần trong ngoặc đơn.

    Hàm này giả định rằng `x[0]` là một chuỗi có định dạng:
        "Phần1 - Phần2 (thông tin thêm)"
    Nó sẽ:
    1. Tách chuỗi theo dấu " - " thành hai phần: a và b
    2. Loại bỏ phần trong ngoặc đơn của phần b nếu có
    3. Trả về hai giá trị dưới dạng Series

    Tham số
    ----------
    x : list
        Danh sách chứa ít nhất một chuỗi ở vị trí `x[0]` theo định dạng nêu trên.

    Kết quả trả về
    ----------
    pandas.Series
        Series gồm hai phần tử:
        - index 0: phần a (chuỗi trước dấu " - ")
        - index 1: phần b đã loại bỏ thông tin trong ngoặc đơn (nếu có)
"""
    a , b = x[0].split(" - ")
    try: 
        b = b.split(" (")[0]
    except:
        None
    return pd.Series([a, b])
````

````python
graphicCard[["Loại card", "Tên card"]] = graphicCard["Card màn hình"].apply(classify)
````

##### kiến thức về nhân GPU
> Phổ biến nhất là nhân CUDA của NVIDIA, thường sẽ rất nhiều (2k - 10k) \
> Các nhân của GPU thì thường sẽ đo bằng card rời phiên bản tiêu chuẩn, chứ k phải phiên bản được làm dành cho laptop \
> Vì vậy đa số k hiển thị nhân GPU, còn nhân GPU tôi đang lấy ra của Appple \
> Apple sử dụng 1 con chipset đó là M series cho cả xử lý đồ họa và tính toán, vì vậy nhân đó là nhân của CPU Apple 

````python
def get_core(text):
    """
    Trích xuất số nhân GPU từ chuỗi tên card đồ họa.

    Hàm này tìm kiếm trong chuỗi đầu vào mẫu số nguyên theo định dạng:
        "<số> nhân GPU"
    và trả về số nhân dưới dạng chuỗi. Nếu không tìm thấy, trả về None.

    Tham số
    ----------
    text : str
        Chuỗi chứa tên card đồ họa, ví dụ: "NVIDIA GeForce RTX 3060 3584 nhân GPU".

    Kết quả trả về
    ----------
    str hoặc None
        Số nhân GPU được trích xuất dưới dạng chuỗi, hoặc None nếu không tìm thấy.
    """
    if isinstance(text, str):
        match = re.search(r'(\d+)\s+nhân GPU', text)
        if match:
            return match.group(1)
    return None

graphicCard["Số nhân GPU"] = graphicCard["Tên card"].apply(get_core)
````

##### logic xử lý đoạn dưới (handle macbook)
> e.g: ["Card tích hợp - xx nhân GPU"; "Card tích hợp"]
##### Vấn đề
Khi tách chuỗi, `Tên card` có thể trở thành `"xx nhân GPU"` thay vì tên thực tế của GPU.

##### Các bước xử lý

1. *Xoá các giá trị thừa*
   - Những hàng mà `"Số nhân GPU"` khác NaN → đặt `"Tên card"` = NaN.

2. *Xác định vị trí (index) của GPU và CPU*
   - Lọc các dòng có `"Brand" = Apple"`.
   - Lấy `idx` của GPU và CPU tương ứng.

3. *Ghép thông tin MacBook*
   - So sánh `idx` của GPU và CPU.
   - Nếu trùng → trả về chuỗi:
     ```
     f"M series + {CPU.model}"
     ```
##### Kết quả
- Các MacBook sẽ có thông tin GPU đúng gắn với CPU.
- Tránh tình trạng `"Tên card"` bị ghi nhầm là `"xx nhân GPU"`.
