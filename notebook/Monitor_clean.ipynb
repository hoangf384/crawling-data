```python
def colorness(x):
    """
    Chuẩn hoá độ phủ màu màn hình.

    - Nếu x là list và chứa cả '100% sRGB' và '100% DCI-P3',
      thì trả về '100% DCI-P3'.
    - Nếu không, trả về giá trị ban đầu.

    Args:
        x (list | str): Dữ liệu độ phủ màu.

    Returns:
        str | list: Giá trị chuẩn hoá độ phủ màu.
    """
    if isinstance(x, list):
        if "100% sRGB" in x and "100% DCI-P3" in x:
            return "100% DCI-P3"
        return x 
    return x  

monitor["Độ phủ màu"] = monitor["Độ phủ màu"].apply(colorness)
```

```python
def classify_resolution(row):
    """
    Phân loại độ phân giải màn hình theo chuẩn Full HD, 2K, 2.8K, 3K, 4K, ...
    Nếu là MacBook thì giữ nguyên giá trị cũ.
    
    row: một dòng DataFrame, cần có row['Độ phân giải'] và row['brand']
    """
    res = row['Độ phân giải']
    
    # Nếu là Apple MacBook, giữ nguyên
    if df.loc[row.name, "brand"] == "MacBook":
        return res
    
    # Lấy độ phân giải dạng số pixel ngang x dọc
    match = re.search(r"(\d{3,4})\s*[xX×]\s*(\d{3,4})", res)
    if match:
        width = int(match.group(1))
        height = int(match.group(2))
        # Phân loại dựa trên chiều dọc (height)
        if height <= 1080:
            return "Full HD"
        elif 1080 < height <= 1440:
            return "2K"
        elif 1440 < height <= 1800:
            return "2.8K"
        elif 1800 < height <= 2000:
            return "3K"
        elif 2000 < height <= 2500:
            return "2.5K"
        elif 2160 < height <= 2500:
            return "4K"
        else:
            return f"{width} x {height}"
    else:
        # Nếu không match được số pixel, giữ nguyên
        return res
```

```python
monitor["resolution_class"] = monitor.apply(classify_resolution, axis=1)
monitor["resolution_class"].value_counts()
```

```python
def extract_nits(x):
    """
    Lấy số nằm trước 'nits' trong x.
    Nếu x là list, trả về phần tử đầu tiên tìm thấy.
    Nếu không tìm thấy, trả về None hoặc x giữ nguyên.
    """
    # Nếu x là list, lặp qua các phần tử
    if isinstance(x, list):
        for item in x:
            match = re.search(r'(\d+(?:\.\d+)?)\s*nits', str(item), re.IGNORECASE)
            if match:
                return f"{match.group(1)} nits"
    
        return None  # không tìm thấy gì trong list
    else:
        # x là string
        match = re.search(r'(\d+(?:\.\d+)?)\s*nits', str(x), re.IGNORECASE)
        if match:
            return f"{match.group(1)} nits"
        else:
            return None
```

```python
monitor["Độ sáng"] = monitor["Công nghệ màn hình"].apply(extract_nits)
```
